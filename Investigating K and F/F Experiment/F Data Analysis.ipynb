Importing relevant libraries for data wrangling

```python
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_baseline = pd.DataFrame()
cols_list_baseline = [1]
for i in range(8, (60*7 + 2), 7):  # Strategy is in every 7th column starting at the 6th column for 60 traders
    cols_list_baseline.append(i)

df_baseline_row_times = []
for i in range(0, 2588400, 3600):
    df_baseline_row_times.append(i)

df_baseline_col_names = ["time"]
for i in range(30):
    if i < 10:
        df_baseline_col_names.append("B0" + str(i))
    else:
        df_baseline_col_names.append("B" + str(i))

for i in range(30):
    if i < 10:
        df_baseline_col_names.append("S0" + str(i))
    else:
        df_baseline_col_names.append("S" + str(i))


for i in range(10):
    #df_k6_i = pd.DataFrame()
    df_baseline_i_list = []
    df_baseline_i_list.append(pd.read_csv("PRDE_Baseline_" + str(i) + "_strats.csv", header=None, usecols=cols_list_baseline))
    df_baseline_i = pd.concat(df_baseline_i_list)

    df_baseline_i.columns = df_baseline_col_names
    df_baseline_i.set_index("time", inplace=True)
    df_baseline[f"Total PPS {i}"] = df_baseline_i.sum(axis=1)

df_baseline["Mean PPS"] = df_baseline.sum(axis=1) / 10
df_baseline
```

```python
# Statistical Testing
baseline_means = df_baseline.mean(axis=0)
stats_df = pd.DataFrame(baseline_means, columns=["F=0.8 (Baseline)"])

F12_means = df_F12.mean(axis=0)
F16_means = df_F16.mean(axis=0)
F20_means = df_F20.mean(axis=0)
stats_df["F=1.2"] = F12_means
stats_df["F=1.6"] = F16_means
stats_df["F=2.0"] = F20_means


new_labels = ["Run 0 Mean PPS", "Run 1 Mean PPS", "Run 2 Mean PPS", "Run 3 Mean PPS", "Run 4 Mean PPS",
              "Run 5 Mean PPS", "Run 6 Mean PPS", "Run 7 Mean PPS", "Run 8 Mean PPS", "Run 9 Mean PPS"]


stats_df.drop(index="Mean PPS", inplace=True)
stats_df.index = new_labels
stats_df
```

```python
# Visualising with a boxplot
BWP = sns.boxplot(data=stats_df)
BWP.set_ylabel("Mean PPS Per Run", fontsize=14)
BWP.set_title("Mean PPS Per Run: PRDE with Varying k and Constant F(=0.8)", fontsize=16)
plt.savefig("Box_and_Whisker_F.png")
plt.show()
```

```python
# Testing for Normality with Shapiro-Wilk test
for col in stats_df.columns:
    SWstatistic, SWpvalue = stats.shapiro(stats_df[col])
    print(f"SW: {col} data is Normally Distributed") if SWpvalue > 0.05 else print(f"SW: {col} data is Not Normally Distributed")

# Confirming Shapiro-Wilk results with Kolmogorov-Smirnov test
for col in stats_df.columns:
    normal_data = (stats_df[col] - stats_df[col].mean())/stats_df[col].std()
    KSstatistic, KSpvalue = stats.kstest(normal_data, "norm")
    print(f"KS: {col} data is Normally Distributed") if KSpvalue > 0.05 else print(f"KS: {col} data is Not Normally Distributed")
```

```python
# Data is confirmed for approximate normality -> Anova Statistical Testing
statistic, pvalue = stats.f_oneway(stats_df["F=0.8 (Baseline)"], stats_df["F=1.2"], stats_df["F=1.6"], stats_df["F=2.0"])
stmnt1 = f"ANOVA: (p= {pvalue} < 0.05). Null hypothesis rejected, distributions have a different population mean."
stmnt2 = f"ANOVA: (p= {pvalue} > 0.05). Null hypothesis not rejected, distributions have the same population mean."
print(stmnt1) if pvalue < 0.05 else print(stmnt2)
```

```python
# Confirmed that populations have a different mean -> t-tests on individual groups
pairs = [["F=0.8 (Baseline)", "F=1.2"], ["F=1.2", "F=1.6"], ["F=1.6", "F=2.0"]]
for i in range(3):
    statistic, pvalue = stats.ttest_ind(stats_df[pairs[i][0]], stats_df[pairs[i][1]])
    stmnt1 = f"T-test: {pairs[i][0]}/{pairs[i][1]} (pvalue= {pvalue} < 0.05). Reject that these samples have identical mean"
    stmnt2 = f"T-test: {pairs[i][0]}/{pairs[i][1]} (pvalue= {pvalue} > 0.05). Can't reject that these have identical mean"
    print(stmnt1) if pvalue < 0.05 else print(stmnt2)
```

All hypothesis that the samples are of identical mean can be rejected. This means that either longer duration testing is needed (50->100 days) to see if the mean PPS of these parameters start to converge, OR, testing of higher F values needs doing in the future to determine the limit of how aggressive PRDE can be before it experiences a loss of performance. Either way it can be seen that F has more of an impact on performance than k.
